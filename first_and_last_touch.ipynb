```python
import csv
import json
import sys
import os
import pandas as pd
```

```python
data = pd.read_csv("Datasets/Path_UTMsource_session_0_price.csv")
#drop the row which Path_Length is 1
data = data[data.Path_Length != 1]

#choose the row that Conversion is 1
data = data[data.Conversion == 1]

#count the proportion of last element in the path
data['last'] = data['Cleaned_Path'].apply(lambda x: x.strip("[]").split(',')[-1].strip("' '"))
data['first'] = data['Cleaned_Path'].apply(lambda x: x.strip("[]").split(',')[0].strip("' '"))
```

```python

last_ = [  'Facebook', 'google', 'LINE', 'affiliate', 'ig', 'PersonalizedMarketing', 'System_Inform','(direct)','POYABuy','Audience' ]
first_list = [  'Facebook', 'google', 'LINE', 'affiliate', 'ig', 'PersonalizedMarketing', 'System_Inform','(direct)','POYABuy','Audience' ]

data['last'] = data['last'].apply(lambda x: x if x in last_ else 'others')
data['first'] = data['first'].apply(lambda x: x if x in first_list else 'others')

print("\n")

print("Last touch attribution after others:")
print(data['last'].value_counts(normalize=True))

print("\n")

print("First touch attribution after others:")
print(data['first'].value_counts(normalize=True))
```

```python
#calculate sum of TotalSalesAmount of every channel
data['TotalSalesAmount'] = data['TotalSalesAmount'].apply(lambda x: x if x != 'NULL' else 0)
data['TotalSalesAmount'] = data['TotalSalesAmount'].astype(float)
last = data.groupby('last')['TotalSalesAmount'].sum()
first = data.groupby('first')['TotalSalesAmount'].sum()

print("\n")
print("Last touch attribution after others:")
print(sorted(last.items(), key=lambda x: x[1], reverse=True))

print("\n")
print("First touch attribution after others:")
print(sorted(first.items(), key=lambda x: x[1], reverse=True))

#save the result to csv
last.to_csv("last_touch.csv")
first.to_csv("first_touch.csv")

#calculate the proportion of TotalSalesAmount of every channel
last = last / last.sum()
first = first / first.sum()

print("\n")
print("Proportion of Last touch attribution after others:")
print(sorted(last.items(), key=lambda x: x[1], reverse=True))

print("\n")
print("Proportion of First touch attribution after others:")
print(sorted(first.items(), key=lambda x: x[1], reverse=True))

#save the result to csv
last.to_csv("last_touch_proportion_2.csv")
first.to_csv("first_touch_proportion_2.csv")
```

```python
#recalculate the proportion compare to sum without others
last = last.drop('others')
first = first.drop('others')

last_sum = last.sum()
first_sum = first.sum()

last = last / last_sum
first = first / first_sum

print("\n")
print("Proportion of Last touch attribution after others:")
print(sorted(last.items(), key=lambda x: x[1], reverse=True))

print("\n")
print("Proportion of First touch attribution after others:")
print(sorted(first.items(), key=lambda x: x[1], reverse=True))

#save the result to csv
last.to_csv("last_touch_proportion_ads_0.csv")
first.to_csv("first_touch_proportion_ads_0.csv")
```
